```python
#slicing-> [start:end:step]
sentence = "Hello, Python Programming!"
print(sentence[0:5])#Hello
print(sentence[7:13])#Python
print(sentence[:5])#Hello
print(sentence[7:])#Python Programming!
print(sentence[::2])#Hlo yhn Pormig!
print(sentence[::-1])#!gnimmargorP nohtyP ,olleH
```
