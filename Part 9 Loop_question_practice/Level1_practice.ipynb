```python
#1. Sum of first n natural numbers.

n = int(input("Enter a number: "))
sum = 0
for i in range(1, n+1):
    sum += i

print(sum)
    
```

```python
#2. Print even numbers from 1 to N

n = int(input("Enter a number: "))

for i in range(1, n+1):
    if i % 2 == 0:
        print(i, end = " ")
        
```

```python
#3. Print odd numbers from 1 to N
n = int(input("Enter a number"))
for i in range(1, n+1):
    if i % 2 != 0:
        print(i, end = " ")
```

```python
#4. print table of a given number

n = int(input("Enter anumber: "))
for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")
```

```python
#5.count digit
num = int(input("Enter a number: "))
count = 0
while (num > 0):
    num = num // 10
    count += 1
print(count)
    
    
    
```
