```python
#6. Reverse the given number
n = int(input("Enter a number: "))
rev = 0
while(n > 0):
    a = n % 10
    rev = rev * 10 + a
    n = n // 10
print(rev)
    
     
```

```python
#7. Sum og digits
n = int(input("Enter a number: "))
sum = 0
while(n > 0):
    a = n % 10 
    sum += a
    n = n // 10

print(sum)
```

```python
#8.product of digits
n = int(input("Enter a number: "))
product = 1
while (n > 0):
    a = n % 10 
    product *= a
    n //= 10
print(product)
```

```python
#9.largest digits in number
n = int(input("Enter a number: "))
max = 0
while(n > 0):
    digit = n % 10
    if digit > max:
        max = digit
    n //= 10
print(max)
```

```python
#10. Smallest digit in number
n = int(input("Enter a number: "))
min = n % 10
while n!=0:
    digit = n % 10
    if digit < min:
        min = digit

    n //= 10
print(min)
```
