```python
def arithmetic(num1, num2):
    add = num1 + num2
    sub = num1 - num2
    mul = num1 * num2
    return add, sub, mul

val_1 = int(input("Enter a number: "))
val_2 = int(input("Enter other number: "))

result1, result2, result3 = arithmetic(val_1, val_2)
print(f"Addition of {val_1} and {val_2} is {result1}")
print(f"Subtraction of {val_1} and {val_2} is {result2}")
print(f"Multiplication of {val_1} and {val_2} is {result3}")



```

```python
#types of argument
def add(a, b):
    return a+b


#positional argument: passing arguments in order their position.
result = add(13, 45)

```

```python
#default argument: assign default values to one or more argument, 
# #then we dont need to pass the values of that argument which we are making default.
# if the value is not passed then it uses the default values assigned
# if we pass the value then, the default values will be overriden

def add(a, b=10):
    return(a+b)

print(add(12, 23))
print(add(12))
```

```python
#keyword arguments:way of passing the argument while calling the function.
#order does not matter
def add(a, b=10, c=10):
    return a+b+c

#result = add(10, 20, 30)
result = add(10, c=50)
print(result)
```

```python
#variable length positional arguments : (0 to n)
# called *args -> can also take a arguments
# *args Python me ek special syntax hai jo function ko 
# variable number of positional arguments accept karne deta hai.
# writing args is not mandatory it is only a standard which we use to take
#technically anything can be wriiten on place of args
#*args exhaust the all positional arguments
def add(*args):
    print(args)
    print(type(args))#tuple
    return sum(args)
add(12, 34)
```

```python
# variable length keyword argument
# **kwargs -> kwargs is not mandatory its a standard only to write
# 0 to n number of arguments
# kwargs are stored in a dictionary
# if we dont pass any argument then, it will give a empty dictionary
def func(**kwargs):
    print(kwargs)
    print(type(kwargs))
    
func(x = 10, y = 40)

```

```python
#**marks should always be in last we cant give it before sid, sname

def student_details(sid, sname, **marks):
    if len(marks)==0:
        print(f"{sname} did not attend the exam.")
    else:
        percent = sum(marks.values())/len(marks)
        print(f"{sname} with id {sid} secured {percent}%.")
        
student_details(101, "Roshan", sub1=78.5, sub2=81.0, sub3=98.5)
student_details(102, "Rohan", sub1 = 34, sub2 = 48, sub4= 89, sub3=34)
        
    
```

```python
#**args should be last the last argument in function definition
def student_details(sid, sname, *extra, **marks):
    if len(marks)==0:
        print(f"{sname} did not attend the exam.")
    else:
        percent = sum(marks.values())/len(marks)
        print(f"{sname} with id {sid} secured {percent}%.")
    print(f"{sname} does {extra}")
        
student_details(101, "Roshan", "football", sub1=78.5, sub2=81.0, sub3=98.5)
student_details(102, "Rohan", "Tennis", "debate", sub1 = 34, sub2 = 48, sub4= 89, sub3=34)
        
    
```
