```python
#filter() takes the first argument as a function
# sequence as a second argument
#filter(function, sequence)
#filter() function keeps the element of sequence for which the filter() function is true

seq = [1, 2, 3, 4]
odd = lambda x: True if x % 2 != 0 else False
filtered_output = filter(odd, seq)
print(filtered_output)
print(f"Odd numbers in the above sequence are: {list(filtered_output)}")

```

```python
seq = [2, 45, 6, 98, 34]
filtered_output = filter(lambda x: True if x % 2 == 0 else False
, seq)
print(filtered_output)
print(f"Even numbers in the above sequence are: {list(filtered_output)}")
```

```python
#map() -> takes function as first argument and sequence as a second argument
#map(function, sequence)
#map object is created at true
#map function keeps wharever output we get
seq = [2, 45, 6, 98, 34]
mapped_output = map(lambda x: True if x % 2 == 0 else False, seq)
print(mapped_output)
print(list(mapped_output))
```

```python
seq = [2, 45, 6, 98, 34]
mapped_output = map(lambda x: x**2, seq)
print(mapped_output)
print(list(mapped_output))
```
