```python
'''
Recursion is a function in which a function cals itself till a certain condition is not met.
factorial of n => (n-1)*(n-2)...*2*1
'''
def fact(num):
    factorial =1
    while(num > 1):
        factorial *= num
        num -= 1
    return factorial

fact(3)
    
```

```python
def fact_rec(num):
    if num == 1:
        return 1
    else:
        factorial = num * fact_rec(num-1)
        return factorial
    
print(fact_rec(4))
```
