```python
# we should write docstring of function defore the logic, then only help function work.
# if we write logic before the doc string then it will not work properly.

def divide(num1, num2):
    """
    num1: A number to be divided(Numerator)
    num2: number that divides num1 (Denominator)
    return: float
    """
    if num2==0:
        return "Cnnot divide as denominator is 0!"
    else:
        result = num1 / num2
        return result

#print(help(divide))
help(divide)
print(divide(10, 5))
print(divide(10, 0)) #: division by zero

    
```
