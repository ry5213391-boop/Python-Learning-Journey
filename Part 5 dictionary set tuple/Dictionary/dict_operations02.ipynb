```python
#allowed keys -> strings, int, float, boolean, tuples
#not allowed keys -> list, set, dict because
#list, set, dict are mutable data structures
# keys of a dictionary can  only be a mutable data type
# values can be a any data type.
```

```python
student1 = {"id": 1001, "name": "john", "marks": [89.5, 71.5, 81.0]}
print(student1)
print(student1["marks"][1]) #fetching values from list
```

```python
student1 = {"id": 1001, "name": "john", "marks": {"eng":89.5, "math": 71.5, "biology": 81.0}}
print(student1)
print(student1["marks"]["eng"])
```

```python
#we cant fetch the keys in the values
# we can fetch all the keys of the dictionary
# key() function
print(student1.keys(), type(student1.keys()))#dict_keys(['id', 'name', 'marks'])
```

```python
# values() function
print(student1.values(), type(student1.values()))

```

```python
# items() function it is used to fetch the keys and values together
print(student1.items(), type(student1.items()))
```
