```python
# dict -> comma separated  key value pairs enclosed with in the curly brackets {}
# {key1:value1, key2:value2....}
groceries = {'milk':60, "biscuit": 10, "rice": 90, "bread":30}
print(groceries)
print(type(groceries))
```

```python
# len()
print(len(groceries)) #4
```

```python
# indexing -> not suported in dictionary
#print(groceries[0]) #KeyError: 0
```

```python
# to fetch values we can use keys
print(groceries['milk']) # 60
```

```python
# dictionaries are mutable

groceries['milk'] = 65
print(groceries)
```

```python
#print(groceries['eggs'])
#KeyError: 'eggs'
```

```python
groceries['eggs'] = 10 
# adds new key-value pair to dictionary
groceries['bread'] = 100 # update the value of the key
print(groceries)
```
