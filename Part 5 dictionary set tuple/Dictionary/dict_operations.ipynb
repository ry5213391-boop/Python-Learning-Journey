```python
s1 = {"math": 80.5, "eng": 76.0, "phy": 89.0}
print(s1)
```

```python
print(s1["phy"])
#print(s1["chem"]) -> KeyError: 'chem'
```

```python
# get()
print(s1.get("math"))
print(s1.get("chem"))#None
print(s1.get("chem",40.0))
print(s1)
```

```python
emp1 = {"id": 1001, "name":"john", "salary": 10000}
print(emp1.get("phone", 23243345)) 
# if key is not present then default value is returned
print(emp1.get('id',23243345))
# if kry is present then the actual value of that key is returned
```

```python
# membership operator => in, not in
print(1001 in emp1) #False
# membership operator does not work with values
print("name" in emp1) # True
# membership operator will work with keys
```

```python
#update()

sem1_marks = {"math": 78.5, "eng": 71.0, "phy": 86.5}
sem2_marks = {"chem": 81.5, "bio": 90.5}

sem1_marks.update(sem2_marks)
print(sem1_marks)
```

```python
thing_1 = {"milk": 29, "rice": 65, "biscuit": 20}
thing_2 = {'rice': 110, "bread": 30}
thing_1.update(thing_2)
print(thing_1)


```

```python
# pop() -> to delete
thing_1.pop("milk") # with milk price also gets deleted
print(thing_1)
```

```python
thing_1 = {"milk": 29, "rice": 65, "biscuit": 20, "milk":78}
print(thing_1)
# the most recent value will be printed when attached to the same key
```
