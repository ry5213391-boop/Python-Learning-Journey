```python
#concatenation, repetition
student_detail1 = (1001, "john")
student_detail2 = (78.5, 91.0, 83.5, 79.5)

student_details = student_detail1 + student_detail2
print(student_details)
```

```python
#repetition-> *
t1 = ("class 5", 5000)
print(t1*3)
```

```python
#membership operator -> in, not in
print(91.0 in student_details) #True
print(91.0 not in student_details) # False
```

```python
# read operations on tuples
#count, index, min, max, sum
t1 = (10, 4, 2, 9, 0, 31, 2, 4, 1, 0, 1)
#tuple.count(element)
print(t1.count(1))
```

```python
#index
#tuple.index(element)
print(t1.index(4))
print(t1.index(1)) # first occurence
# print(t1.index(40)) # ValueError: tuple.index(x): x not in tuple
```

```python
#min max sum
print(min(t1))
print(max(t1))
print(sum(t1))
```
