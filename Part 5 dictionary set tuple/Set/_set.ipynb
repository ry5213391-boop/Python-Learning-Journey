```python
# sets are  collection of items
# sets are non sequential collection
# comma separated items enclosed with in curly brackets
# cannot have indexing with sets
# indexing and slicing both are not allowed
# duplicate elements are not allowed
# anything that is unique
```

```python
set1 = {10, "python", 2.5}
print(set1)
print(type(set1))
```

```python
l1 = [10, 2, 3.4, 30, 10]
print(l1, type(l1))

s1 = {10, 2, 4.4, 30, 10}
print(s1, type(s1))
#duplicate elements are removed automatically
```
