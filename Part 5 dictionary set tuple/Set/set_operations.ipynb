```python
nums = {1, 3, 2, 0, -1}

#Membership operator -> in, not in
print(0 in nums)
print(10 not in nums)
```

```python
weekdays = ("Mon", "tue", "wed", "thurs", "fri", "sat", "sun")
weekdays = set(weekdays)
print(weekdays)
# sets are mutable but 
# order can change randomly
# non- sequential and unorder data structure
```

```python
set1 = {1, 0, -1}
print(id(set1))
# add()-> it will add randomly at any where
set1.add(5)
print(set1)
print(id(set1))
set1.add(1)# doesnot give error
print(set1) # no difference in set
```

```python
#remove()
s1 = {1, 2, 3, 4, 5}
print(s1)
s1.remove(1)
print(s1)
#s1.remove(10) -> KeyError: 10 means we cant remove it
print(s1)
```

```python
# discard() -> will delete the elements
s2 = {1, 3, 6, 0, -4}
s2.discard(0)
print(s2)
s2.discard(20)# it will not give error
print(s2)# no difference in set
```
