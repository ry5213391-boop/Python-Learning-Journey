```python
# these concepts are only applied for mutable data types.



# copy function of copy module is used to create shallow copies.
```

```python
import copy

# shallow copy means only outer list have same memory addresses and inner list dont have the memory addresses.
l1 = [1, 2.4, [10, 20, 30], "Python"]
print(id(l1))
l2 = copy.copy(l1) #shallow copy
print(l2)
print(id(l2))


# both l1 and l2 will have different memory addresses but
# the memory addresses of inner elements of both l1 and l2 will have same memory addresses
```

```python
l1[0] = 100 # when we change directly then it will not changed.
l1[2][0] = 50 # when we change internally then it wil be change
print(f"l1 -> {l1}", id(l1))
print(f"l2 -> {l2}", id(l2))
```

```python
# deep copy -> when inner memory addresses are also become different

l1 = [1, 2.4, [10, 20, 30], "Python"]
l2 = copy.deepcopy(l1)
l1[0] = 5
#l1[2][0] = 50
print(l1, id(l1), id(l1[0]))
print(l2, id(l2), id(l2[0]))

```

```python
d1 = {"id":1111, "name": "John", "marks": {"eng": 71.5, "maths": 91.5, "bio":80.0}}
#deep copy
d2 = copy.deepcopy(d1)
d1["name"] = "Roshan"
# directly so not any change in d1
print(f"d1 -> {d1}", id(d1))
print(f"d2 -> {d2}", id(d2))
```
