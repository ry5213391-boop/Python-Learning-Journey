```python
s1 = "Hello World"
print(len(s1))
```

```python
#indexing
print("First char:", s1[0])
print("Last char:", s1[-1])
```

```python
#slicing = is a technique where it enables a programmer to access or fetch a part of the string.
#[start:end:step]
print(s1[2:7:1])
#start will be included
#end will be excluded, when not given then included
```
