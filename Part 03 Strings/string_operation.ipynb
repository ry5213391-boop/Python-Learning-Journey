```python
#replace() function
s1 = "We are learning Python"
print(s1)
print(s1.replace("Python", "Java"))
#syntax-> string.replace(old_string, new_string)
#existing string will not change
#replace function will remove the every existing character present in the string
#  and which we want to remove and add new string  
print(s1)
```

```python
print(s1.replace("e","E"))
```

```python
print(s1.replace("e","E",1))
#now it only chnages the first character of the string
#if we want to change n number of occurences then we will give as 1, 2 as so on
```

```python
print(s1.replace("e","E",5))
#now it become independent that as the number of character which we want to remove
# gets completed it will automatically replace existing characters either we increse the number
# or not increase the number.
```
