```python
#append() in list
#adds an item to the end of the list

fruits = ["Mango", "Apple", "Orange"]
#list.append(item)
print(fruits.append("Banana"))#None
# but in string if we print something using this then we get the output
# Beacuse we cannot change the existing string, any function that we will use with the string
# then we will always get a new string
print(fruits)
```

```python
# lists are mutable-> means it can change the original list when any modification is 
# done in the original list
```

```python
fruits = ["Mango", "Apple", "Orange"]
print(fruits.append("Banana"))#None
print(fruits)
#append func does not return the updated list
#it does not create a new list
#existing list got changed and this is called mutability
```

```python
#insert() -> adds an element before  the specified index, 
s1 = "ROSHAN"
s2 = "YADAV"
#print(s1.insert(s2)) #AttributeError: 'str' object has no attribute 'insert'
```

```python
#insert() -> in lists
#syntax: list.insert(index, item)
fruits = ["Mango", "Apple", "Orange"]
fruits.insert(-1, "Banana")
print(fruits)
```

```python
#extend() -> add no. of elements at the end of the list
li = ["apple", "mango", "orange"]
print(li)
#li.append("banana", "grapes") #TypeError: list.append() takes exactly one argument (2 given)
li.extend(["banana", "grapes"])

print(li)
```

```python
#remove()->removes certain elements from the list
fruits = ["apple", "banana", "orange"]
print(fruits)
fruits.remove("apple")
#fruits.remove("apple", "orange")
#TypeError: list.remove() takes exactly one argument (2 given)

print(fruits)
```

```python
#pop()-> does not take the element, it removes the element from the specified index and return also
```

```python
fruit = ["apple", "mango", "mango", "apple"]
print(fruit.pop(2))
print(fruit.pop(-1))
print(fruit)
```

```python
li = ["ai", "ml", "dl", "gai"]
print(li)
li.pop() # if index is not given then it will automatically removes the last element from the list
print(li)
```

```python
#reverse() = reversed the list and changes the list itself.
#doesnot gives a new list, the list itself gets changed
days_of_week = ["sun", "mon", "tue","wed", "thu", "fri", "sat"]
print(days_of_week)
days_of_week.reverse()
print(days_of_week)
```

```python
nums = [4, 9, 0, 1, 7, 5]
print(nums)
#sort() -> it sorts the list in ascending order
nums.sort()
print("Sorted list:",nums)
```

```python
nums.sort(reverse=True) #arrange in descending order
print(nums)
```

```python
#count()->count the particular element is a list that how many times that occured and return also
numbers = [1, 0, 4, 7, 8, 3, 2, 9, 3, 0, 3, 8, 0]
#print(numbers.count(0)) -> 3
#numbers.count(0)
print("The list is:",(numbers))
item_to_count = int(input("Enter the number to be counted from the above list:"))
c = numbers.count(item_to_count)
print(f"Occurences of {item_to_count} is {c}")
```

```python
#membership operator in list: in
lang = ["python", "java", "c++", "c", "perl"]
print("python" in lang)# True
print("ruby" not in lang) #True
print("javascript" in lang) #False
```
