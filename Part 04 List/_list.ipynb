```python
days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
#in list also indexing works
#indexing = [0,1,2,3,4,5,6,7]
#every data is associated with a index
print(days_of_week[0])
print(days_of_week[-1])
print(days_of_week[0:])
print(days_of_week[:])
print(days_of_week[0:6:2])
print(days_of_week[-1:6])
print(len(days_of_week)) #7
print(days_of_week[-1:-100])
print(days_of_week[2:-8])
print(days_of_week[1:-1:2])
#print(days_of_week[8]) #IndexError: list index out of range
#this indexerror also occur in strings
```
