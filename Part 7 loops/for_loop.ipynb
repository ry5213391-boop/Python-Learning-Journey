```python
#for loop
# it is an iterator based loop which steps through the items of a collection
#(list, tuple, dictionary, set, str), and a block of code repeatedly
#for a number of times equal to the items/elements of that collection.
```

```python
percent = [85.5, 81, 86, 83.5]
print(percent[0])
print(percent[2])
```

```python
for p in percent:
    print(p)
```

```python
s1 = "Hello world"

for char in s1:
    print(char, end=" ")
print("Roshan Yadav")
```
