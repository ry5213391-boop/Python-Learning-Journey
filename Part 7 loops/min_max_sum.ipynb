```python
scores = [2, 45, 102, 4, 9, 12, 45, 90, 1, 0, 1]

total = sum(scores)
print(total)
```

```python
#highest score

highest = scores[0] #assume that the first score is the highest

for score in scores:
    if highest < score:
        highest = score

print(highest)
```

```python
high = max(scores)
print(high)
```

```python
lowest = scores[0]
for score in scores:
    if score < lowest:
        lowest = score
    
print(lowest)
```

```python
low = min(scores)
print(low)
```
