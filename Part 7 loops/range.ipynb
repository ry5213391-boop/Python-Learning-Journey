```python
#range() -> built in fun in python
#it is used to generate a sequence of integers
# range(start, stop, step)
# stop is included in the generation
```

```python
for i in range(1, 10, 1):
    print(i, end = " ")
```

```python
# reverse order = 20 to 10

for i in range(20, 10, -1):
    print(i, end = " ")
    
```

```python
for i in range(10, 0, -1):
    print(i)
```

```python
#range(start, stop) => step = 1 by default
# range(stop) => 0 to stop-1 with a step of 1, start = 0 by default
```

```python
groceries = ["salt", "milk", "sugar"]

for index in range(len(groceries)):
    print(index)
```

```python
profits = [9, 11, 6, 10]

for index in range(len(profits)):
    q = index + 1
    print(f"Profits for quarter {q} is {profits[index]}")
```
