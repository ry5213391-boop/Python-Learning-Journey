```python
age = float(input("What is your age?"))

if age >= 18:
    print("Congrats! you are an adult. You can now cast vote!!!.")
```

```python
# if - else
age = float(input("What is your age?"))

if age >= 18:
    print("Congrats! you are an adult. You can now cast vote!!!.")
else:
    print("No you are not eligible to vote for now.")
```

```python
num = float(input("Enter a number: "))
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
num = float(input("Enter a number: "))
if (num > 0):
    print("Positive number.")
elif (num == 0):
    print("Zero is neither positive nor negative")
    
else:
    print("Negative number.")
```

```python
# marks grade

marks = float(input("Enter marks: "))
if (marks >= 90):
    print(f"Grade: A")
elif (89 >= marks >= 80):
    print("Grade B")
elif (79 >= marks >= 70):
    print("Grade C")
elif (69 >= marks >= 60):
    print("Grade D")
else:
    print("Grade F")
```

```python
# Nested if
marks = float(input("Enter marks: "))
if marks >= 60:
    print("Congrats, you has passed the exam.")
    if (marks >= 90):
        print(f"Grade: A")
    elif (89 >= marks >= 80):
        print("Grade B")
    elif (79 >= marks >= 70):
        print("Grade C")
    else: 
        print("Grade D")
    
else:
    print("You have failed, study hard next time.")
```
